```python
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
import pandas as pd
import datetime
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from collections import Counter
plt.rcParams.update({'font.size': 22})
import sys
```

```python
temps = pd.read_csv('templatedets.dat', delim_whitespace=True, names=['No', 'Date', 'Time', 'Lat', 'Lon', 'Dep', 'Mag', 'Coef', 'Times_MAD', 'Reference'])
new = pd.read_csv('newdets.dat', delim_whitespace=True, names=['No', 'Date', 'Time', 'Lat', 'Lon', 'Dep', 'Mag', 'Coef', 'Times_MAD', 'Reference'])

temps.head()

tc = list(temps['Coef'])
tc = [float(c) for c in tc]
nc = list(new['Coef'])
nc = [float(c) for c in nc]

cumtem = []
cc = []
cumnew = []
for c in np.arange(0, 1.1, 0.01):
    ct = sum(map(lambda x: x<=c, tc))
    cn = sum(map(lambda x: x<=c, nc))
    cumtem.append(ct)
    cc.append(c)
    cumnew.append(cn)

    
#print(cumtem)
#print(cumnew)





```

```python
cumtem.reverse()

cumnew.reverse()

#print(cumtem)
#print(cumnew)
cc.reverse()
fig, ax = plt.subplots(figsize=(8,8))
plt.scatter(cc, cumtem, label='self detections', color='magenta')
plt.scatter(cc, cumnew, label='new detections', color='slateblue')
plt.xlim(max(cc), min(cc))
plt.ylabel('Número de eventos detectados')
plt.xlabel('CC')
plt.legend()
plt.title('Valor del CC en función del número de detecciones')
```
